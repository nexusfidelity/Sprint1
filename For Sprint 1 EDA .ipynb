```python
data_1_male = data_1[['School ID','Kinder Male','Grade 1 Male','Grade 2 Male','Grade 3 Male','Grade 4 Male','Grade 5 Male','Grade 6 Male']]
```

```python
data_1_female = data_1[['School ID','Kinder Female','Grade 1 Female','Grade 2 Female','Grade 3 Female','Grade 4 Female','Grade 5 Female','Grade 6 Female']]
```

```python
data_urbanity_1 = data_schools[['school.id','school.urban','school.curricularclass']]
```

```python
data_urbanity_2 = data_urbanity_1[(data_urbanity_1['school.curricularclass']=='Elementary')]
```

```python
elem_urbanity = data_urbanity_2[['school.id','school.urban']]
```

```python
elem_urbanity['school.id']= pd.to_numeric(elem_urbanity['school.id'],errors='coerce')
```

```python
elem_urbanity['school.id']=elem_urbanity['school.id'].astype(float)
```

```python
df_male = pd.merge(elem_urbanity,data_1_male, left_on='school.id', right_on='School ID',how='inner')
```

```python
df_female = pd.merge(elem_urbanity,data_1_female, left_on='school.id', right_on='School ID',how='inner')
```

```python
df_male_urban = df_male [(df_male['school.urban']=='Urban')]
df_male_p_urban = df_male [(df_male['school.urban']=='Partially Urban')]
df_male_rural= df_male [(df_male['school.urban']=='Rural')]

df_female_urban = df_female [(df_female['school.urban']=='Urban')]
df_female_p_urban = df_female [(df_female['school.urban']=='Partially Urban')]
df_female_rural= df_female [(df_female['school.urban']=='Rural')]
```

```python
df_male_p_urban = df_male_p_urban.drop(columns=['school.urban','school.id'])
df_male_rural = df_male_rural.drop(columns=['school.urban','school.id'])
df_female_p_urban = df_female_p_urban.drop(columns=['school.urban','school.id'])
df_female_rural = df_female_rural.drop(columns=['school.urban','school.id'])
df_female_urban = df_female_urban.drop(columns=['school.urban','school.id'])
```

```python
df_male_urban= df_male_urban[['School ID', 'Kinder Male', 'Grade 1 Male', 'Grade 2 Male',
       'Grade 3 Male', 'Grade 4 Male', 'Grade 5 Male', 'Grade 6 Male']]
df_male_p_urban= df_male_p_urban[['School ID', 'Kinder Male', 'Grade 1 Male', 'Grade 2 Male',
       'Grade 3 Male', 'Grade 4 Male', 'Grade 5 Male', 'Grade 6 Male']]
df_male_rural= df_male_rural[['School ID', 'Kinder Male', 'Grade 1 Male', 'Grade 2 Male',
       'Grade 3 Male', 'Grade 4 Male', 'Grade 5 Male', 'Grade 6 Male']]

df_female_urban= df_female_urban[['School ID', 'Kinder Female', 'Grade 1 Female', 'Grade 2 Female',
       'Grade 3 Female', 'Grade 4 Female', 'Grade 5 Female', 'Grade 6 Female']]
df_female_p_urban= df_female_p_urban[['School ID', 'Kinder Female', 'Grade 1 Female', 'Grade 2 Female',
       'Grade 3 Female', 'Grade 4 Female', 'Grade 5 Female', 'Grade 6 Female']]
df_female_rural= df_female_rural[['School ID', 'Kinder Female', 'Grade 1 Female', 'Grade 2 Female',
       'Grade 3 Female', 'Grade 4 Female', 'Grade 5 Female', 'Grade 6 Female']]
```

```python
df_m_p_urban_sum = df_male_p_urban.sum().to_frame().drop(labels=['School ID'],axis=0,inplace=False)
df_m_rural_sum = df_male_rural.sum().to_frame().drop(labels=['School ID'],axis=0,inplace=False)
df_f_rural_sum = df_female_rural.sum().to_frame().drop(labels=['School ID'],axis=0,inplace=False)
df_f_urban_sum = df_female_urban.sum().to_frame().drop(labels=['School ID'],axis=0,inplace=False)
df_f_p_urban_sum = df_female_p_urban.sum().to_frame().drop(labels=['School ID'],axis=0,inplace=False)
df_m_urban_sum = df_male_urban.sum().to_frame().drop(labels=['School ID'],axis=0,inplace=False)
```

```python
df_m_urban_sum.plot(kind='bar')
df_m_p_urban_sum.plot(kind='bar')
df_m_rural_sum.plot(kind='bar')

```

```python
ax = df_m_urban_sum.plot()
df_f_urban_sum.plot(ax=ax)

plt.show()

ax = df_m_p_urban_sum.plot()
df_f_p_urban_sum.plot(ax=ax)

ax = df_m_rural_sum.plot()
df_f_rural_sum.plot(ax=ax)
```
