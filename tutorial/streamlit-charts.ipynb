```python
import datetime
import folium
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import streamlit as st

from folium.features import DivIcon
from spacy_streamlit import visualize_ner
from streamlit_folium import folium_static

st.set_option('deprecation.showPyplotGlobalUse', False)
```

### Column and Bar Chart

```python
df = pd.DataFrame({'countries': ['USA', 'Myanmar', 'Philippines', 'India', 'Greece', 'Singapore', 'China',
                                         'Malaysia', 'UK', 'Colombia'],
                           'count': [10, 2, 4, 6, 8, 3, 5, 6, 8, 9]})
```

```python
fig = plt.figure(figsize=(8,5))
ax1 = sns.barplot(data=df, x='countries', y='count',  ci=None, color='#5dade2')
ax1.set_title('Sample Bar Chart')
ax1.set(xlabel='Countries', ylabel='Count')        
plt.xticks(rotation=45)
plt.show()  ### Please remove this line when running on streamlit

st.pyplot()
```

```python
fig = plt.figure(figsize=(8,5))
ax1 = sns.barplot(data=df, x='count', y='countries',  orient = 'h', ci=None, color='#f1948a')
ax1.set_title('Sample Bar Chart')
ax1.set(xlabel='Countries', ylabel='Count')        
plt.show()  ### Please remove this line when running on streamlit

st.pyplot()
```

### Line Chart

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
may_flights = flights.query("month == 'May'")

fig = plt.figure(figsize=(8, 5))
ax1= sns.lineplot(data=may_flights, x="year", y="passengers", color='#5dade2')
ax1.set_title('Sample Line Chart')
ax1.set(xlabel='Year', ylabel='Passengers')        
plt.xticks(rotation=45)
plt.show()  ### Please remove this line when running on streamlit

st.pyplot()
```

### Scatterplot

```python
tips = sns.load_dataset("tips")

fig_1 = plt.figure(figsize=(8,5))
ax1 = sns.scatterplot(data=tips, x="total_bill", y="tip", color='#5dade2')
ax1.set_title('Total Bills per Day')
ax1.set(xlabel='Day', ylabel='Total Bills')    
plt.show() ### Please remove this line when running on streamlit

st.pyplot()
```

### Bubble Chart

```python
from gapminder import gapminder 

fig_1 = plt.figure(figsize=(8, 5))
data = gapminder.loc[gapminder.year == 2007]
ax1 = sns.scatterplot(data=data, x="gdpPercap", y="lifeExp", size="pop", legend=False, sizes=(20, 2000))
ax1.set_title('Relationship between life expectancy (y) and gdp per capita (x)')
ax1.set(xlabel='GDP Per Capita', ylabel='Life Expectancy') 
plt.show() ### Please remove this line when running on streamlit

st.pyplot()
```

### Histogram

```python
penguins = sns.load_dataset("penguins")

fig_1 = plt.figure(figsize=(8, 5))
ax1 = sns.histplot(data=penguins, x="flipper_length_mm", color='#f1948a')
ax1.set_title('Flipper_length_mm Distribution')
ax1.set(xlabel='flipper_length_mm', ylabel='Count') 
plt.show() ### Please remove this line when running on streamlit

st.pyplot()
```

### Boxplot

```python
tips = sns.load_dataset("tips")

fig_1 = plt.figure(figsize=(8, 5))
ax1 = sns.boxplot(x="day", y="total_bill", data=tips, palette="Blues")
ax1.set_title('Total Bills Per Day')
ax1.set(xlabel='day', ylabel='Total Bill') 
plt.show() ### Please remove this line when running on streamlit

st.pyplot()
```
