```python
# 1st feature Student Ratio
final_data_na["student_ratio"] = final_data_na["school.enrollment"] / final_data_na["teachers_total"]
```

```python
# 2nd feature Rooms Ratio
final_data_na["rooms_ratio"] = final_data_na["school.enrollment"] / final_data_na["rooms_total"]
```

```python
final_data_na["school.mooe"] = final_data_na[" school.mooe "].str.replace("," , "")
```

```python
final_data_na[["school.mooe"]] = final_data_na["school.mooe"].astype("float")
```

```python
final_data_na = final_data_na.drop(columns=[' school.mooe '])
```

```python
# 3rd feature MOOE per student Ratio
final_data_na["mooe_per_student"] = final_data_na["school.mooe"] / final_data_na["school.enrollment"]
```
